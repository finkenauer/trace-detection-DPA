```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
km = pd.read_csv('km_tam.csv', header = None, names = 'k')
km.describe()
```

```python
fz = pd.read_csv('fuzzy_tam.csv', header = None, names = 'f')
fz.describe()
```

```python
import seaborn as sns
km.k.sort_values().plot(kind='hist', bins = 15)
```

```python
fz.f.sort_values().plot(kind='hist', bins = 15)
```

```python
sns.kdeplot(km.k)
```

```python
 sns.kdeplot(fz.f)
```

```python
plt.boxplot(km.k)
```

```python
plt.boxplot(fz['f'])
```

```python
plt.boxplot(fz['f']), plt.boxplot(km['k'])
```

```python
tam = pd.read_csv('tam.csv')
tam.head()
```

```python
plt.boxplot(tam.km), plt.boxplot(tam.fz)
```

```python
plt.boxplot(tam.km~tam.fz)
```

```python
tam.boxplot(sym='+')
```

```python
fig, ax = plt.subplots()
tam.km.plot(kind='box', ax=ax)

```
